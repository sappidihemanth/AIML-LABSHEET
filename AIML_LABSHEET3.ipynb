```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.title(['Apples','Oranges']);
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')

plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.title(['Apples','Oranges']);
```

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,mew=10,alpha=0.5)

plt.xlabel('year')
plt.ylabel('yield')

plt.title('Crop yields in Kanto')
plt.title(['Apples','Oranges']);
```

```python
plt.plot(year,apples,'s--b')
plt.plot(year,oranges,'o--r')
plt.xlabel('years')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,oranges,'or')
plt.title('yield of oranges')
```

```python
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title("Bar Chart")
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.scatter(x,y)

plt.legend("A")
plt.xlabel('X')
plt.ylabel('Y')

plt.title("Scatter plot")
plt.show()
```

```python
x = [1,2,3,4]
e = (0.1,0,0,0)
plt.pie(x, explode = e)

plt.title("Pie Chart")
plt.show()
```
